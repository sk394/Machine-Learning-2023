```python
import tkinter as tk

class TicTacToe:
    def __init__(self):
        self.root = tk.Tk()
        self.root.title("Tic Tac Toe")
        self.root.geometry("200x200")

        self.board = [[' ' for _ in range(3)] for _ in range(3)]
        self.player = 'X'
        self.buttons = []

        for i in range(3):
            row = []
            for j in range(3):
                button = tk.Button(self.root, width=5, height=2, command=lambda x=i, y=j: self.play(x, y))
                button.grid(row=i, column=j)
                row.append(button)
            self.buttons.append(row)

    def play(self, x, y):
        if self.board[x][y] == ' ':
            self.board[x][y] = self.player
            self.buttons[x][y].config(text=self.player)
            self.check_win()
            if self.player == 'X':
                self.player = 'O'
            else:
                self.player = 'X'
                move = get_best_move(self.board)
                if move is not None:
                    x, y = move
                    self.board[x][y] = self.player
                    self.buttons[x][y].config(text=self.player)
                    self.check_win()
                    self.player = 'X'

    def check_win(self):
        """Check if the game is over and display the winner if there is one."""
        result = evaluate(self.board)
        if result is not None:
            if result == 1:
                message = "X wins!"
            elif result == -1:
                message = "O wins!"
            else:
                message = "It's a tie!"
            tk.messagebox.showinfo("Game Over", message)
            self.root.destroy()

    def run(self):
        self.root.mainloop()
        
def evaluate(board):
    """
    Evaluate the current state of the board and return a score for the AI player.
    A positive score indicates a win for the AI player, a negative score indicates a win for the opponent,
    and a score of 0 indicates a tie.
    """
    # Check for a win for the AI player
    if has_won(board, 'X'):
        return 1
    # Check for a win for the opponent
    elif has_won(board, 'O'):
        return -1
    # Check for a tie
    elif is_full(board):
        return 0
    else:
        return None

def has_won(board, player):
    """Check if the given player has won the game."""
    for i in range(3):
        # Check rows
        if board[i][0] == player and board[i][1] == player and board[i][2] == player:
            return True
        # Check columns
        if board[0][i] == player and board[1][i] == player and board[2][i] == player:
            return True
    # Check diagonals
    if board[0][0] == player and board[1][1] == player and board[2][2] == player:
        return True
    if board[0][2] == player and board[1][1] == player and board[2][0] == player:
        return True
    return False

def is_full(board):
    """Check if the board is full."""
    for i in range(3):
        for j in range(3):
            if board[i][j] == ' ':
                return False
    return True

def generate_moves(board, player):
    """Generate all possible moves from the current state of the board."""
    moves = []
    for i in range(3):
        for j in range(3):
            if board[i][j] == ' ':
                moves.append((i, j))
    return moves

def minimax(board, player):
    """
    Recursively evaluate all possible moves and return the best move for the AI player.
    """
    result = evaluate(board)
    if result is not None:
        return None, result

    best_move = None
    best_score = -2

    for move in generate_moves(board, player):
        x, y = move
        board[x][y] = player
        _, score = minimax(board, 'O' if player == 'X' else 'X')
        board[x][y] = ' '
        score *= -1

        if score > best_score:
            best_move = move
            best_score = score

    return best_move, best_score

def get_best_move(board):
    """Return the best move for the AI player."""
    move, _ = minimax(board, 'X')
    return move
if __name__ == '__main__':
    game = TicTacToe()
    game.run()
```
