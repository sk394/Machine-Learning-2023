```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
X = cancer.data
y = cancer.target
#no need to scale data for decision trees
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0)
print(X_train.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(max_depth=3,random_state=0)
dt.fit(X_train,y_train)
```

```python
print(dt.score(X_train,y_train))
print(dt.score(X_test,y_test))
```

```python
train_acc=[]
test_acc=[]
for i in range(1,10):
    dt = DecisionTreeClassifier(max_depth=i,random_state=0)
    dt.fit(X_train,y_train)
    train_acc.append(dt.score(X_train,y_train))
    test_acc.append(dt.score(X_test,y_test))
```

```python
test_acc
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,10),train_acc,'r')
plt.plot(range(1,10),test_acc,'b')
```

```python
dt.feature_importances_
```

```python
plt.barh(range(X.shape[1]),dt.feature_importances_)
plt.yticks(range(X.shape[1]),cancer.feature_names)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=70,random_state=0)
rf.fit(X_train,y_train)
```

```python
print(rf.score(X_train,y_train))
print(rf.score(X_test,y_test))
```

```python
train_acc=[]
test_acc=[]
for i in range(1,15):
    rf =RandomForestClassifier(n_estimators=i*10,random_state=0) 
    rf.fit(X_train,y_train)
    train_acc.append(rf.score(X_train,y_train))
    test_acc.append(rf.score(X_test,y_test))
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,15),train_acc,'r')
plt.plot(range(1,15),test_acc,'b')
```

```python
plt.barh(range(X.shape[1]),rf.feature_importances_)
plt.yticks(range(X.shape[1]),cancer.feature_names)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gb = GradientBoostingClassifier(n_estimators=50,learning_rate=0.2, max_depth=1,random_state=0)
gb.fit(X_train,y_train)
```

```python
print(gb.score(X_train,y_train))
print(gb.score(X_test,y_test))
```
