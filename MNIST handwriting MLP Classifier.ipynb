```python
#Classify MNIST handwriting dataset
from sklearn.datasets import fetch_openml
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import MinMaxScaler

#load the dataset
handwriting = fetch_openml('mnist_784')
X = handwriting.data[:1000]
y = handwriting.target[:1000]
print("success")
```

```python
scaler = MinMaxScaler()
scaler.fit(X)
X_scale = scaler.transform(X)

#split the dataset into training and testing
X_train, X_test, y_train, y_test = train_test_split(X_scale, y, test_size=0.3, random_state=0)

#define classifier
mlp = MLPClassifier()

#define hyperparameters to tune the model
param_grid = {
    'hidden_layer_sizes':[(50,),(100,),(100,100)],
    'activation':['relu', 'logistic','tanh'],
    'solver':['adam', 'sgd'],
    'alpha':[0.001,0.01,0.1],
    'learning_rate':['constant', 'adaptive']
}

#use GridSearch to find best param
grid_search = GridSearchCV(mlp, param_grid = param_grid, cv=5)
grid_search.fit(X_train, y_train)

print("Best hyperparameters: ",grid_search.best_params_ )

#evaluate the accuracy
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy on test data: {accuracy}')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles

# Generate dataset
X, y = make_circles(n_samples=200, noise=0.04, factor=0.6, random_state=0)

# Plot original dataset
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.RdBu)
plt.title("Original Dataset")
plt.show()

# Define KernelSVM class
class KernelSVM:
    def __init__(self, gamma=1):
        self.gamma = gamma
        self.alpha = None
        self.support_vectors = None

    def fit(self, X, y, learning_rate=0.001, epochs=100):
        n_samples, n_features = X.shape

        # Initialize alpha values
        self.alpha = np.zeros(n_samples)

        # Calculate kernel matrix
        K = np.zeros((n_samples, n_samples))
        for i in range(n_samples):
            for j in range(n_samples):
                K[i,j] = (1 + self.gamma * np.dot(X[i], X[j])) ** 2

        # Gradient descent
        for epoch in range(epochs):
            for i in range(n_samples):
                # Calculate margin
                margin = 0
                for j in range(n_samples):
                    margin += self.alpha[j] * y[j] * K[i,j]
                margin *= y[i]

                # Update alpha
                if margin < 1:
                    self.alpha[i] += learning_rate

        # Find support vectors
        self.support_vectors = self.alpha > 1e-5

    def predict(self, X):
        n_samples, n_features = X.shape

        # Calculate kernel matrix
        K = np.zeros((n_samples, len(self.support_vectors)))
        for i in range(n_samples):
            for j in range(len(self.support_vectors)):
                K[i,j] = (1 + self.gamma * np.dot(X[i], X[self.support_vectors[j]])) ** 2

        # Predict class
        y_pred = np.dot(K, self.alpha[self.support_vectors] * y[self.support_vectors])
        return np.sign(y_pred)

    def score(self, X, y):
        y_pred = self.predict(X)
        return np.mean(y_pred == y)

# Transform dataset using second-order polynomial kernel
svm = KernelSVM(gamma=1)
svm.fit(X, y)
X_transformed = np.zeros((X.shape[0], 3))
for i in range(X.shape[0]):
    X_transformed[i, 0] = X[i, 0]
    X_transformed[i, 1] = X[i, 1]
    X_transformed[i, 2] = (1 + svm.gamma * np.dot(X[i], X[i])) ** 2

# Plot transformed dataset
plt.scatter(X_transformed[:, 1], X_transformed[:, 2], c=y, cmap=plt.cm.RdBu)
plt.title("Transformed Dataset")
plt.show()
```
