```python
#problem1: load online dry dataset and have a quick check of the dataset
import pandas as pd
import requests
import zipfile
import io

#get/load the data from website
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00602/DryBeanDataset.zip'
response = requests.get(url)

# Extract the excel file from the zip file
zipfile = zipfile.ZipFile(io.BytesIO(response.content))
csv_file = zipfile.open('DryBeanDataset/Dry_Bean_Dataset.xlsx')

frame = pd.read_excel(csv_file)

#quick check of the data
frame.head()
```

```python
#get the info from the dataset
frame.info()
```

```python
#get the mean/median/max of the columns data
frame.describe()
```

```python
#boxplot of area with class
import matplotlib.pyplot as plt

#separate the data by two columns 0th and 16th
result = frame.iloc[:, [0, 16]]

#names to the columns-represented->
result.columns = ['Area', 'Class']

#create a boxplot charcaterized by Class-->easy to visualize(qualitative and quantitative variable)
result.boxplot(column='Area', by='Class')
plt.xlabel("Class")
plt.ylabel("Bean Area")
plt.show()   #note(from the visualization, we can say-> Bombay class beans have larger area and more IQR than others)
```
