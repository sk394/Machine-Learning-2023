```python
#Use K-nearest neighbors method to classify the dataset generated by the code below
from sklearn.datasets import make_classification
X,y= make_classification(n_samples=1000, n_features=2, n_redundant=0, n_informative=2, n_clusters_per_class=1,n_classes=4)

#plotting data
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
#evaluating maxm accuracy using K-neighbors
from sklearn.neighbors import KNeighborsClassifier
knn= KNeighborsClassifier(n_neighbors=3)
```

```python

#training model

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X,y,test_size=0.25, random_state=20)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

#model evaluation
accuracy= knn.score(X_test, y_test) *100

print("The accuracy of the prediction is {}% ".format(accuracy))
```

```python
import matplotlib.pyplot as plt
plt.figure(1)
plt.plot(y_pred, accuracy)
```

```python
#problem 1- #Use K-nearest neighbors method to classify the dataset generated by the code below
from sklearn.datasets import make_classification
X,y= make_classification(n_samples=1000, n_features=2, n_redundant=0, n_informative=2, n_clusters_per_class=1,n_classes=4)

#plotting data
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
#evaluating/classifying maxm accuracy using K_neighbors
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt   #for plotting the graph n_neighbors VS accuracy
accuracy_array =[]
#store the consequent accuracy values in an array
for i in range(1,10):
    knn = KNeighborsClassifier(n_neighbors=i)
    #training/testing model using model selection
    from sklearn.model_selection import train_test_split
    X_train,X_test, y_train, y_test = train_test_split(X,y, test_size=0.50, random_state = 25)
    """"
    This line of code is using the train_test_split function from the 
    sklearn.model_selection library to split the dataset into two sets: a training set and a testing set.

The train_test_split function takes four arguments:

X: is the dataset that we want to split
y: is the target variable.
test_size: is the proportion of the dataset to include in the test split.
It is set to 0.50, which means that half of the dataset will be used for testing and the other half for training.
random_state: is the seed used by the random number generator. Setting the 
random_state to 25 makes the split reproducible, so if the code is run again
with the same random_state, it will produce the same training and testing sets.
The function returns four variables:

X_train: represents the training set of the independent variable
X_test: represents the testing set of the independent variable
y_train: represents the training set of the dependent variable
y_test: represents the testing set of the dependent variable
Therefore, this line of code is splitting the dataset X and y into two sets
X_train and X_test and y_train and y_test, respectively. The test_size is 
set to 0.50, meaning that half of the dataset is used for testing and half
for training. The random_state is set to 25 to ensure reproducibility.
    """
    knn.fit(X_train, y_train)
    y_pred= knn.predict(X_train)
    accuracy = knn.score(X_test, y_test)
#print the accuracy percentage
    print("The accuracy is {}% when n_neighbor value is {}".format(round(accuracy*100,2),i))
    accuracy_array.append(accuracy) #store/add each accuracy values into the array
#plot the n_neighbors VS accuracy graph
plt.figure(figsize=(6,7))
plt.plot(range(1,10),accuracy_array)
plt.xlabel("n_neighbors")
plt.ylabel("accuracy")
plt.show()
```
